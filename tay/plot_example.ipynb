```python
import numpy as np

N = 100 # map size

data = np.load('/home/amitbou/One_Shot_Path_Planning/tay/amit_compressed.npz')
inp = data['inp']
target = data['target']
db_size = inp.shape[0]
del data
```

```python
import matplotlib.pyplot as plt

example_idx = np.random.randint(0, 50)
input_example = inp[example_idx, :, :, 0]  
end_point_example = inp[example_idx, :, :, 1]  
target_example = target[example_idx, :, :] 

plt.figure(figsize=(10, 5))

plt.subplot(1, 3, 1)
plt.title('Input: Obstacles')
plt.imshow(input_example, cmap='gray')
plt.colorbar()

plt.subplot(1, 3, 2)
plt.title('Input: End Point')
plt.imshow(end_point_example, cmap='gray')
plt.colorbar()

plt.subplot(1, 3, 3)
plt.title('Target: Distance Map')
plt.imshow(target_example, cmap='hot')
plt.colorbar()

plt.show()
```
